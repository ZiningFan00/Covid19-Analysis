```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

### Read Datasets

```python
total_df = pd.read_excel (r'total.xlsx')
state_df = pd.read_excel (r'state.xlsx')
```

### Total Doses

```python
total_df.columns = ['State', 'a','b','c','d']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 13))

# fig.title('Total Doses vs 18+ Doses per 100k')

sns.barplot(ax=axes[0], y = total_df.iloc[:,0], 
            x = total_df.iloc[:,2], 
            data = total_df,
            order = total_df.sort_values('b',ascending = False).State)
sns.barplot(ax=axes[1], y = total_df.iloc[:,0], 
            x = total_df.iloc[:,4], 
            data = total_df,
            order = total_df.sort_values('d',ascending = False).State)

axes[0].set_title('Total Doses per 100k')
axes[1].set_title('18+ Doses per 100k')

plt.tight_layout()
```

### State Doses

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=state_df, x="date", 
             y="people_fully_vaccinated_per_hundred",
             hue = "location",
             palette = "Set2")
plt.legend(loc='best')
plt.title("People fully vaccinated per 100")
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=state_df, x="date", 
             y="total_vaccinations_per_hundred",
             hue = "location",
             palette = "Set2")
plt.legend(loc='best')
plt.title("Total Vaccinations per 100")
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=state_df, x="date", 
             y="people_vaccinated_per_hundred",
             hue = "location",
             palette = "Set2")
plt.legend(loc='best')
plt.title("People vaccinated per 100")
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 15))
sns.lineplot(data=state_df, x="date", 
             y="daily_vaccinations_per_million",
             hue = "location",
             palette = "Set2")
plt.legend(loc='best')
plt.title("Daily Vaccinations per 1M")
plt.tight_layout()
```

### Take California as an example

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

fig.suptitle('California Data')

sns.lineplot(ax=axes[0, 0], data=state_df.loc[state_df['location'] == 'California'], x="date", 
             y="people_fully_vaccinated_per_hundred",
             palette = "Set2")
sns.lineplot(ax=axes[0, 1], data=state_df.loc[state_df['location'] == 'California'], x="date", 
             y="total_vaccinations_per_hundred",
             palette = "Set2")
sns.lineplot(ax=axes[1, 0], data=state_df.loc[state_df['location'] == 'California'], x="date", 
             y="people_vaccinated_per_hundred",
             palette = "Set2")
sns.lineplot(ax=axes[1, 1], data=state_df.loc[state_df['location'] == 'California'], x="date", 
             y="daily_vaccinations_per_million",
             palette = "Set2")

axes[0,0].set_title('People Fully Vaccinated per 100')
axes[0,1].set_title('Total Vaccinations per 100')
axes[1,0].set_title('People Vaccinated per 100')
axes[1,1].set_title('Daily Vaccinations per 1M')

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

fig.suptitle('New York Data')

sns.lineplot(ax=axes[0, 0], data=state_df.loc[state_df['location'] == 'New York State'], x="date", 
             y="people_fully_vaccinated_per_hundred",
             palette = "Set2")
sns.lineplot(ax=axes[0, 1], data=state_df.loc[state_df['location'] == 'New York State'], x="date", 
             y="total_vaccinations_per_hundred",
             palette = "Set2")
sns.lineplot(ax=axes[1, 0], data=state_df.loc[state_df['location'] == 'New York State'], x="date", 
             y="people_vaccinated_per_hundred",
             palette = "Set2")
sns.lineplot(ax=axes[1, 1], data=state_df.loc[state_df['location'] == 'New York State'], x="date", 
             y="daily_vaccinations_per_million",
             palette = "Set2")

axes[0,0].set_title('People Fully Vaccinated per 100')
axes[0,1].set_title('Total Vaccinations per 100')
axes[1,0].set_title('People Vaccinated per 100')
axes[1,1].set_title('Daily Vaccinations per 1M')

plt.tight_layout()
```
